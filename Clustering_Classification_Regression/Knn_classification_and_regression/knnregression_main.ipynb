Generate N = 300 samples for training and 100 for testing sets. The first 2 columns in train should be the 2D Gaussian random data and the last column should be the target value. Use L2-norm vectors compute distance between different samples. Try out different k = 1, 2, 3, 5, 10, 20, 50, 100 to test your k-NN classifier.
```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
mean_0 = np.array([1,0])
variance_0 = np.array([[1, 0.75], [0.75, 1]])
x_train = np.random.multivariate_normal(mean_0, variance_0, 300)
mean_1 = 0
variance_1 = 0.5
noise_1 = np.random.normal(mean_1, variance_1, 300)
y_train=[]
for i in range(len(noise_1)):
    y1 = 2*x_train[i][0] + x_train[i][1] + noise_1[i]
    y_train.append(y1)
X = np.insert(x_train, 2, values = y_train, axis = 1)
```

```python
x_test = np.random.multivariate_normal(mean_0, variance_0, 100)
noise_2 = np.random.normal(mean_1, variance_1, 100)
y_test=[]
for i in range(len(noise_2)):
    y2 = 2*x_test[i][0] + x_test[i][1] + noise_2[i]
    y_test.append(y2)
test = np.insert(x_test, 2, values = y_test, axis = 1)
```

```python
from knnregression import myknnregress, getAccuracy
k=3
print('Predicted value: ' + repr(myknnregress(X, test[99], k)))		
```

```python
plt.plot(x_train[:, 0], x_train[:, 1], 'bo')
plt.title("Training Data")
plt.show()

plt.plot(x_test[:, 0], x_test[:, 1], 'go')
plt.title("Testing Data")
plt.show()
```

```python
k = [1,2,3,4,5,10,20,50,100]
scores = []
j = 0
for i in k:
    print("K: ", i)
    scores.append(getAccuracy(X, test, i)) 
    print("***********")
```
